```python
import random
import numpy
import scipy 
import matplotlib

M=0
x=[]
y=[]

for n in range(200) :
    a = random.uniform(-1.0, 1.0)
    b = random.uniform(-1.0, 1.0)
    x.append(a)
    y.append(b)

for n in range(200):
    if x[n]**2 + y[n]**2 <= 1:
        M += 1

pi = 4 * M/(n+1)
#print(n, M, pi)
print(pi)
```

```python
import random
import numpy
import scipy 
import matplotlib
import math
from math import sqrt

pi=[]

for j in range(100):
    
    M=0
    x=[]
    y=[]

    for n in range(1000):
        a = random.uniform(-1.0, 1.0)
        b = random.uniform(-1.0, 1.0)
        x.append(a)
        y.append(b)
        
        if x[n]**2 + y[n]**2 <= 1:
            M += 1

    pig = 4 * M/(n+1)
    pi.append(pig)
    

pi_media=sum(pi)/len(pi)
print(pi_media)

num=[]

for i in range(100):
    c=(pi[i]-pi_media)**2
    num.append(c)

arg=sum(num)/(len(num)-1)

sigma = sqrt(arg)

print(sigma)

```
